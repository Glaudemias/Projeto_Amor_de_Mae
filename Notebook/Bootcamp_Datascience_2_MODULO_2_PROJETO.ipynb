## 1.1 Cuidado com as Gestantes

Nos ultimos anos tem-se observado uma maior preocupação de se pautar a saúde das mulheres, sobretudo nos paises desenvolvidos. Entretanto, em países em desenvolvimento como é o caso do Brasil, são grandes os problemas relacionados à saúde reprodutiva, destancando-se desde a morte materna ou a complicações durante a gestação (NASCIMENTO; PAIVA;RODRIGUES, 2007).

Nesse sentido se aborda como assistência Pré-Natal(PN) o conjunto de cuidados e procedimentos que se preocupa em assegurar a saúde sa gestante e do bebê, a fim de assegurar a detecção precoce das possiveis complicações da gestação, o tratamento de doenças maternas ja existentes, além de orientação de habitos saudaveis e o preparo da gestante para os momentos do parto e o pós-parto (ou puérperio) (GRANGEIRO; DIÓGENES; MOURA, 2008).

O Sistema Único de Saúde integrado, tenta historicamente promover uma boa politica de garantia dos direitos a saúde do bebê e da gestante. Por fim cabe ressaltar que outro programa responsável pela promoção da segurança de uma boa gestação a mãe, é o Programa Nacional de Imunizações (PNI), cujo objetivo é assegurar a proteção contra doenças transmissíveis em toda a população e nesse caso especifico para o controle do tétano neonatal (ja controlada aqui no Brasil) e a Hepatite B, ambas as doenças são passíveis de imunização prévia e evitáveis (PACHECO, 2011)

## 1.2 Vacinação x Óbitos Fetais

A vacinação de gestantes é considerada prióritária pela Organização Mundial de Saúde (OMS) e representam o primiero ato de proteção materna ao bebê e a sua própria saúde, buscando evitar complicações durante o parto e o desenvolvimento de doenças ao recém nascido.

Segundo a Sociedade Brasileira de Iuminizações (SBIM), os anticorpos maternos são muito importantes para os bebês nos primeiros 12 meses de vida, são eles que irão protege-lo até que este possa ser vacinado e desenvolva suas próprias defesas, evitando assim a ocorrência do aumento do número de óbitos fetais.Segundo o Ministério da Saúde pode-se definir o Óbito fetal como:

> *Por sua vez, a mortalidade fetal partilha com a mortalidade neonatal precoce as
mesmas circunstâncias e etiologia que influenciam o resultado para o feto no final da
gestação e para a criança nas primeiras horas e dias de vida. Os óbitos fetais são também,
em grande parte, considerados potencialmente evitáveis. No entanto, têm sido historicamente negligenciados pelos serviços de saúde, que ainda não incorporaram na sua rotina
de trabalho a análise de sua ocorrência e tampouco destinaram investimentos específicos
para a sua redução (Ministério da Saude, 2009)*

Ainda segundo a SABIM, durante a gravidez o uso de medicamentos é contraindicados, portanto cabe a vacina evitar problemas como:

* Malformações congênitas e danos da gestação(aborto, nascimento prematuro, entre outros).

* Gestantes são consideradas um grupo de alto risco para influenza (gripe), assim como os bebês e quando essa ocorre nos primeiros meses de vida, pode ser motivo de hospitalização e quadros respiratórios graves.

* No Brasil praticamente 100% das mortes por coqueluche são de menores de 6 meses. E a vacinação de gestantes é a principal recomendação para mudar esse cenário.

* A mulher sem saber pode transmitir o vírus da Hepatite B, para seu filho durante a gravidez ou parto e quando isso ocorre cerca de 90% dos bebês chegam a desenvolver a forma crônica da doença

## 1.3 Vacinas Analisadas

vacina-gravida.jpg

Os Imunos (vacinas) que serão analisadas por esse projeto são quatro: **Febre Amarela**, **Hepatite-B**, **DT**, **DTPA**. Quanto as caracteristicas de cada uma dessas vacinas, sabe-se que:

* **Hepatite-B:** A vacinação protege do vírus hepatite B, cause de inflamação no fígado que pode cronificar e, nesse caso, provocar cirrose e câncer.

  Doses = 3 doses

* **Febre Amarela:** A vacinação da Febre Amarela em gestantes é um caso especial, uma vez que o imunizante contra a febre amarela é feito a partir de vírus vivos, que foram atenuados.
Segundo Rosana Richtmann, do **Centro de Imunização do Hospital e Maternidade Santa Joana**, em São Paulo, para a VEJA em 2019. 
>*“As gestantes, assim como os idosos, possuem um risco ligeiramente maior de apresentar efeitos colaterais graves da vacina...Como estão bem fraquinhos, eles dificilmente conseguem vencer o sistema imune da pessoa, que então produz anticorpos capazes de debelar mesmo o vírus normal da doença.”*

  Segundo o manual da [Febre Amarela](https://www.bio.fiocruz.br/en/images/stories/pdfs/bulas/fa/BM_BUL_045_00_V_190702_FA10Nacional.pdf) da Fio Cruz "não há evidências de que a vacinação de uma grávida (por febre amarela) esteja associada a efeitos anormais no feto", podendo gerar complicações apenas na lactação.

  Doses = dose unica  
  
* **DT:** A vacinação protege contra a difteria e o tétano

  Doses = 2
  
  Esquema de doses:Gestante com esquema de doses incompleto para o tétano e a difteria, após receber a vacina dTpa, deve receber duas doses de dT: a primeira, dois meses após a dTpa e, a segunda, 4 a 6 meses após a primeira dT. A recomendação do Ministério da Saúde é que a gestante receba duas doses da vacina contra difteria e tétano e mais uma dose da vacina dTpa. A vacina não contém bactérias vivas, portanto não causa doença.

* **DTPA:** A vacinação protege contra a difteria, o tétano e a coqueluche

  Doses = 3 doses

  Esquema de doses: Deve ser repetida a cada gravidez, a partir da 20ª semana de gestação. A recomendação do Ministério da Saúde é a de que a gestante não adequadamente vacinada contra o tétano e a difteria receba duas doses da vacina dT e mais uma dose da vacina dTpa, com o objetivo de possibilitar a transferência de anticorpos contra a coqueluche para o bebê. No entanto, se ela não se vacinou, deve se imunizar no pós-parto, o mais cedo possível, para evitar transmitir a infecção. A vacina não contém bactérias vivas, portanto não causa doença.
